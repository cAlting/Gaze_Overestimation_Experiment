## Control Flow
- über Bedingungen bzw. Verknüpfung von Bedingungen mit Boolschen Operatoren

```python
a,b,c = 1,2,3

# Evaluiert hier zu True (da ein Oder verwendet wird)
if (a == a) | (b == c):
    print('Test')

# Evaluiert zu False, da ein Und verwendet wird (b ist nicht gleich c)
if (a == a) & (b == c):
    print('Test')    

# Verbinden mehrerer Bedingungen
# Evaluiert hier zu True (da zweimal Oder verwendet wird)
if ((a == a) | (b == c)) | (a==c):
    print('Test')
    
# Evaluiert hier zu False (da dioe beiden öußersten Bedingungen per & verbunden werden)
if ((a == a) | (b == c)) & (a==c):
    print('Test')
```

## Korrelationsplot / Heatmap
- interessante Website: ["Spurious Correlations"](https://www.tylervigen.com/spurious-correlations)

### Einfachste Variante

```python
# Schritt 1: Berechnen der Korrelationen der Variablen
# Spearman, da es sich um ordinale Daten handelt
corr_mat = df_both.corr('spearman')

# Schritt 2: Einfach Heatmap erstellen
# blöd: zu eng, es werden nicht alle Labels angezeigt
sns.heatmap(corr_mat)
```

### Variante 2 (figure size, colormaps)

```python
fig, ax = plt.subplots(figsize=(10,15))

sns.heatmap(corr_mat,
           cmap='bwr') # cmap = color map
```

### Variante 3 (gefilterte Korrelationen)
- manchmal kann es sinnvoll sein, Korrelationen kleiner als ein bestimmter Absolutbetrag nicht zu plotten, um die Übersicht zu verbessern und den Fokus auf relevante Variablen zu legen
- dann ist es sinnvoll, die Korrelationen in der berechneten Korrelationsmatrix zu filtern
- zusätzlich ist es nicht sinnvoll, Korrealtionen auf der Hauptdiagonalen zu plotten, da diese immer = 1 sind

_Vorgehen_:
1. Korrelationen kleiner als bestimmter Cutoff entfernen
2. Korrelationen von Variablen mit sich selbst (Hauptdiagonale) entfernen
3. NA-Werte entfernen

__Anmerkungen zur Heatmap:__
1. _Autokorrelationen_
    - Variablen, die ähnliches messen, korrelieren stark $\rightarrow$ Autokorrelation
2. _Problem: geringe Stichprobengröße bei manchen Variablen_
    - es haben nur wenige Leute Kinder $\rightarrow$ es gehen nur wenige Personen ein, deshalb kleine Stichprobe
    - gleiches gilt für Hilfsangebote
3. _mgl. Lösung:_ Funktion, um nachzuschauen, wie viele und ob es NaNs in Tabelle gibt
    - df.isna()
    - df.notna()
    - (diese mit sum kombinieren)

```python
# Cutoff festlegen
cutoff = 0.5
# Filtern der Korrelationen kleiner als abs(0.4) und == 1
corr_minmax = corr_mat[((corr_mat > cutoff) | (corr_mat < -cutoff)) & (corr_mat != 1)]

# NaN entfernen
# axis = 0: Zeilenweise
# axis = 1: Spaltenweise
corr_minmax_woNA = corr_minmax.dropna(axis=0,
                                     how='all')

corr_minmax_woNA = corr_minmax_woNA.dropna(axis=1,
                                          how='all')

sns.heatmap(data=corr_minmax_woNA,
           cmap='bwr')
```

### Variante 4 (über Funktion)

```python
def plot_corr(df, cmin=-0.4, cmax=0.4, method='pearson'):
    
    # correlation matrix
    corr = df.corr(method)
    
    # correlation cutoff
    corr_minmax = corr[((corr > cmax) | (corr < cmin)) & (corr != 1)]
    
    # drop NaNs
    corr_minmax.dropna(axis=0, how='all', inplace=True)
    corr_minmax.dropna(axis=1, how='all', inplace=True)
    
    # figure erstellen
    fig, ax = plt.subplots(figsize=(30,25))
    
    # Heatmap
    sns.heatmap(corr_minmax,
               cmap='bwr')
    
    # bereinigte Korrelationsmatrix zurückgeben
    return corr_minmax
```

```python
# Prep für die nächsten 2 Werkzeuge
df_d = pd.read_csv('SIDP_cleaned.csv')

# wenn df z.B. 6 Zeilen hätte (= 6 VPs)
numbers = np.arange(1, len(df_d)+1)

# leeres array für VP-IDS
participants = []

# VP-ID-Array füllen
for n in numbers:
    participants.append('VP' + str(n))

# neue Spalte mit VP-IDs erstellen
df_d['Participant'] = participants
df_d.head()
```
